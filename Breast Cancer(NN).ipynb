**#Separating the FEATURES and TARGET**

```python
x = df.drop(columns = 'label', axis = 1)
y = df['label']
```

```python
x
```

```python
y
```

**#Splitting the data into training data and testing data**

```python
# Split features (x) and labels (y) into training and testing sets.
# 80% for training, 20% for testing. random_state=2 ensures consistent results.
# random_state sets the seed for shuffling — ensures the train-test split is the same every time.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x.shape, x_train.shape, x_test.shape
```

**Standardising the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)

x_test_std = scaler.transform(x_test)
```


**BUILDING THE NEURAL NETWORK**

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)       #ensures reproducibility
from tensorflow import keras
```

```python
#setting up the layers of Neural Network
model = keras.Sequential([
    keras.Input(shape=(30,)),
    keras.layers.Flatten(), #input layer....need to be flatten(1D) as u x&y_train are in matrices
    keras.layers.Dense(20, activation='relu'),  #hidden layer..20 neurons...ReLU:helps learn non-linear relationships.
    keras.layers.Dense(2, activation='sigmoid')  #output layer....2 neurons stands for the classes we have here 0 and 1
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',                      
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#Adam optimizer for efficient learning
#Sparse categorical crossentropy for multi-class classification with integer labels
#Accuracy metric to monitor performance
```

```python
# training the Neural Network
# training the Meural Network

history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
#x_train featues other is label
# Uses 10% of training data for validation.
# Runs for 10 epochs....at each epoch loss decreses and accuracy increases
```

**Visualising accuracy and loss**

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

**Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```
